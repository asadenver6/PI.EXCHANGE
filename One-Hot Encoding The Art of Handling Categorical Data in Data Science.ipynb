## Import all the required libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

## Load the dataset, drop Churn column and view the first few rows

```python
df= pd.read_csv('Telco_Churn.csv', skipinitialspace=True)
df= df.drop('Churn', axis=1)
df.head()
```

## Check for missing values in the dataset

```python
for col in df:
    missing_values= df[col].isna().sum()
    if(missing_values>0):
        print('{} {}'.format(col, missing_values))
```

## Extract the categorical features that will be one-hot encoded

```python
categorical_features= df.select_dtypes(include= ['object'])
categorical_features
```

## View the number of categories for the PaymentMethod feature

```python
categorical_features.PaymentMethod.value_counts()
```

## Apply one-hot encoding on the PaymentMethod column

```python
one_hot_encoder = OneHotEncoder(sparse=False)
```

```python
one_hot_encoder.fit_transform(categorical_features[['PaymentMethod']])
```

```python
one_hot_encoder.categories_
```

## Drop the customerID column

```python
 categorical_features= categorical_features.drop('customerID', axis=1)
```

## Apply one-hot encoding on all categorical features 

```python
one_hot_encoder.fit_transform(categorical_features)
```

```python
names= one_hot_encoder.get_feature_names()
names= pd.Series(names)
names= names.apply(lambda x: x.split('_')[1])
```

```python
categorical_features= pd.DataFrame(one_hot_encoder.fit_transform(categorical_features))
categorical_features.columns= names
categorical_features
```
